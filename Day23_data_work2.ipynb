```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('데이터산업기술.csv', encoding='cp949')
pd_data.drop('기술등급별(1)', axis = 1, inplace = True)
```

```python
pd_data['데이터직무별(1)']


```

```python
fig, ax = plt.subplots(2, 4, figsize=(20, 10))
ax = ax.flatten()

for i, column in enumerate(pd_data.columns[1:]):
    ax[i].plot(pd_data['시점'], pd_data['데이터직무별(1)'])
    ax[i].set_title(column)

plt.tight_layout()
plt.show()
```

```python
df_architect1 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 아키텍트']
df_architect2 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 엔지니어']
df_architect3 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 개발자']
df_architect4 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 분석가']
df_architect5 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터베이스 관리자']
df_architect6 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 과학자']
df_architect7 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 컨설턴트']
df_architect8 = pd_data.loc[pd_data['데이터직무별(1)'] == '데이터 기획자']

_, ax = plt.subplots(2, 4, figsize=(20, 10))
ax = ax.flatten()
def plot_lib(x):
    for i, column in enumerate(x.columns[2:]):
        ax[i].plot(x['시점'], x[column])
        ax[i].set_title(column)
    
    
plot_lib(df_architect1)
plot_lib(df_architect2)
plot_lib(df_architect3)
plot_lib(df_architect4)
plot_lib(df_architect5)
plot_lib(df_architect6)
plot_lib(df_architect7)
plt.tight_layout()
plt.show()
```

```python
_, ax = plt.subplots()
ax.plot(df_architect1['시점'], df_architect1.iloc[:, 3])
ax.plot(df_architect1['시점'], df_architect1.iloc[:, 4])
```

```python
from matplotlib import font_manager, rc

font_fname = 'D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

grouped = pd_data.groupby('데이터직무별(1)')

fig, ax = plt.subplots(2, 4, figsize=(20, 10))
ax = ax.flatten()

for i, (name, group) in enumerate(grouped):
    for j, column in enumerate(group.columns[2:]):
        ax[j].plot(group['시점'], group[column], label=name)
        ax[j].set_title(column)

plt.tight_layout()
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(2, 4)
col_names = pd_data.columns[-2:]

for i,n in enumerate(pd_data['데이터직무별(1)'].unique()):
    filter1 = pd_data['데이터직무별(1)'] == n
    xs = pd_data.loc[filter1, '시점']
    ys = pd_data.loc[filter1, col_names]
    axes[i//4, i%4].plot(xs, ys)
    
```
