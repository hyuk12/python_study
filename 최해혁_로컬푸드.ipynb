```python
import requests, csv

url = "https://www.data.go.kr/download/15025455/standard.do?dataType=csv"
response = requests.get(url)

with open("data.csv", "wb") as f:
    f.write(response.content)


items = {}
item_count = {}
providers = {}
with open("data.csv", encoding="cp949") as f:
    reader = csv.DictReader(f)
    for row in reader:
        item_name = row["품목명"].split(" ")[0].split("(")[0].split("+")[0]
        provider_name = row["제공기관명"]
        
        if item_name not in item_count:
            item_count[item_name] = 0
        item_count[item_name] += 1
        
        if item_name not in items:
            items[item_name] = []
        items[item_name].append(provider_name)
        
        
        
item_count_list = [(item_name, count) for item_name, count in item_count.items()]
item_count_list.sort(key=lambda x: x[1], reverse=True)

print('품목명 top10\n')

for item_name, count in item_count_list[:10]:
    print(f"{item_name}: {count}")
print('품목 : 제공기관명\n')
    
for item_name, providers in items.items():
    print(f"{item_name}: {providers}")
```
