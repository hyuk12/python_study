```python
import numpy as np
import pandas as pd

url = 'https://www.data.go.kr/download/15013110/standard.do?dataType=csv'
raw_df = pd.read_csv(url,encoding='cp949')
```

# 전국 평생학습 강좌 표준데이터

- drop: 교육시작시각, 교육 종료 시각, 강좌내용, 운영기관전화번호

- column 별 적절한 dtype지정 --> astype()이용
- (원본값 수정방법, 1. 현재값 업데이트, 2. astype()의 키워드 옵션사용)
- astype() 의 키워드 옵션 사용방법 --> 수치값 np.inta64/ np.float64
- 종류를 구분하는 문자열(label) : category
- 날짜 np.datetime64
- 이외 object

- 전체 DataFrame에 대한 describe()결과보기
- 키보드옵션은 Documentation 참조

- 항목 (column)별 결측치/오류치/이상치 갯수파악하기

```python
raw_df.shape
raw_df= raw_df.drop(['교육시작시각', '교육종료시각', '강좌내용', '운영기관전화번호'], axis = 1)
```

```python
raw_df.dtypes
```

```python
raw_df[['교육시작일자', '교육종료일자']] = raw_df[['교육시작일자', '교육종료일자']].astype(np.datetime64)
```

```python
raw_df.dtypes
```

```python
raw_df[[ '운영요일', '접수시작일자', '접수종료일자', '데이터기준일자', '제공기관코드', '강좌정원수']]
```

```python
raw_df[['접수시작일자', '접수종료일자', '데이터기준일자']] = raw_df[['접수시작일자', '접수종료일자', '데이터기준일자']].astype(np.datetime64)
```

```python
raw_df['운영요일'] = raw_df['운영요일'].astype('category')
raw_df[['교육대상구분', '교육방법구분', '접수방법구분', '선정방법구분', '직업능력개발훈련비지원강좌여부', '학점은행제평가(학점)인정여부', '평생학습계좌제평가인정여부', '제공기관코드']] = raw_df[['교육대상구분', '교육방법구분', '접수방법구분', '선정방법구분', '직업능력개발훈련비지원강좌여부', '학점은행제평가(학점)인정여부', '평생학습계좌제평가인정여부', '제공기관코드']].astype('category')
```
