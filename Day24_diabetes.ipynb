```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('데이터산업.csv', encoding='cp949')
```

```python
pd_data
```

```python
data_processing_and_management = pd_data['데이터 처리 및 관리 솔루션 개발·공급업']
data_construction_and_consulting = pd_data['데이터 구축 및 컨설팅 서비스업']
data_sales_and_provision = pd_data['데이터 판매 및 제공 서비스업']
print(data_processing_and_management)
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))

ax1 = plt.subplot(2,1,1)
ax1.plot(years, data_processing_and_management, 'ro-', label='Data Processing and Management')
ax1.plot(years, data_construction_and_consulting,'g^--', label='Data Construction and Consulting')
ax1.plot(years, data_sales_and_provision, 'bv:', label='Data Sales and Provision')

# Add labels and title
ax1.set_xlabel('Year')
ax1.set_ylabel('Market Size')
ax1.set_title('Market Size Growth by Data Industry')

# Add legend
ax1.legend()

# Show plot


# Pie Chart for 2015
market_sizes1 = [data_processing_and_management[0], data_construction_and_consulting[0], data_sales_and_provision[0]]

industries = ['Data Processing and Management', 'Data Construction and Consulting', 'Data Sales and Provision']
ax2 = plt.subplot(2,2,3)
ax2.pie(market_sizes1, labels=industries, autopct='%1.1f%%')
ax2.set_title('Market Size Composition in 2015')

# Pie Chart for 2021
market_sizes2 = [data_processing_and_management[6], data_construction_and_consulting[6], data_sales_and_provision[6]]

industries = ['Data Processing and Management', 'Data Construction and Consulting', 'Data Sales and Provision']

ax2 = plt.subplot(2,2,4)
ax2.pie(market_sizes2, labels=industries, autopct='%1.1f%%')
ax2.set_title('Market Size Composition in 2021')

# Show plot

plt.show()
```
