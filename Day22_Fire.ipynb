```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url1 = 'https://www.data.go.kr/download/15034538/standard.do?dataType=csv'
url2 = 'https://www.data.go.kr/download/15034536/standard.do?dataType=csv'

raw_data1 = pd.read_csv(url1, encoding='cp949')
raw_data2 = pd.read_csv(url2, encoding='cp949')



pd_hydrant = raw_data1[['시도명', '소재지도로명주소', '위도', '경도']]
pd_parking = raw_data2[['시도명', '소재지도로명주소', '위도', '경도']]

```

```python
print(raw_data1.shape)
print(raw_data2.shape)
```

```python
pd_final = pd.concat((pd_hydrant, pd_parking), axis = 0, keys=['hydrant', 'parking'])
```

```python
pd_final.reset_index(inplace = True)
pd_final.drop('level_1', axis= 1, inplace=True)
cname_list= list(pd_final.columns)
cname_list[0] = '종류'
pd_final.columns = cname_list
pd_final
```

```python
result1 = pd_final.groupby(['시도명', '종류']).size().reset_index(name='counts')
result1
```

```python
result2 = pd_final.groupby(['종류', '시도명']).size().reset_index(name='counts')
result2
```
