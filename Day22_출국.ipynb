```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('내국인출국교통수단.csv', header=1, encoding='cp949')

pd_data[['공항', '항구']] = pd_data[['김해', '부산']]
pd_data.drop(['김해', '부산'], axis = 1, inplace = True)
pd_data.head()
```

## 공항, 항구 년중 월별 출국 인원 plot

- nomalization /  standardization 적용 , plot
- 2015 부터 2022 년까지 추이 보기

```python
def nomal_air_ship(air, ship):
    air_min = air.min()
    air_max = air.max()
    
    ship_min = ship.min()
    ship_max = ship.max()

    pd_data['air_norm'] = air.apply(lambda x: (x- air_min) / (air_max - air_min))
    pd_data['ship_norm'] = ship.apply(lambda x: (x- ship_min) / (ship_max - ship_min))
    return pd_data[['air_norm','ship_norm']]

   

def std_air_ship(air, ship):
    air_mean = air.mean()
    air_std = air.std()
    
    ship_mean = ship.mean()
    ship_std = ship.std()

    pd_data['air_std'] = air.apply(lambda x : (x - air_mean) / air_std)
    pd_data['ship_std'] = ship.apply(lambda x : (x - ship_mean) / ship_std)
    return pd_data[['air_std', 'ship_std']]
    

nomal_air_ship(pd_data['공항'], pd_data['항구'])    
std_air_ship(pd_data['공항'], pd_data['항구'])
    
_, axe = plt.subplots()
# axe.plot(pd_data['공항'])
# axe.plot(pd_data['항구'])
axe.plot(pd_data['air_norm'])
axe.plot(pd_data['ship_norm'])
# axe.plot(pd_data['air_std'])
# axe.plot(pd_data['ship_std'])
```

```python
pd_data = pd.read_csv('내국인출국교통수단2015.csv', header=1, encoding='cp949')

pd_data[['공항', '항구']] = pd_data[['김해', '부산']]
pd_data.drop(['김해', '부산'], axis = 1, inplace = True)
pd_data.head()
```

```python

print(pd_data['시점'].dtype)

pd_data['시점'] = pd_data['시점'].astype('str')
pd_data['시점'] = pd.to_datetime(pd_data['시점'],format = '%Y/%m')

print(pd_data['시점'])
```

```python
year_group = pd_data.groupby(pd_data['시점'].dt.year).sum()
print(pd_data['시점'].dt.year)
```

```python
year_group
```

```python
_, axe = plt.subplots()
axe.plot(year_group['공항'])
axe.plot(year_group['항구'])
```

```python
def year_nomal_air_ship(air, ship):
    air_min = air.min()
    air_max = air.max()

    ship_min = ship.min()
    ship_max = ship.max()
    
    year_group['air_norm'] = air.apply(lambda x: (x- air_min) / (air_max - air_min))
    year_group['ship_norm'] = ship.apply(lambda x: (x- ship_min) / (ship_max - ship_min))
    return year_group[['air_norm', 'ship_norm']]


def year_std_air_ship(air, ship):
    air_mean = air.mean()
    air_std = air.std()
    
    ship_mean = ship.mean()
    ship_std = ship.std()

    year_group['air_std'] = air.apply(lambda x : (x - air_mean) / air_std)
    year_group['ship_std'] = ship.apply(lambda x : (x - ship_mean) / ship_std)
    return year_group[['air_std','ship_std']]
    


year_nomal_air_ship(year_group['공항'], year_group['항구'])
year_std_air_ship(year_group['공항'], year_group['항구'])
```

```python
_, axe = plt.subplots()
axe.plot(year_group['air_norm'])
axe.plot(year_group['ship_norm'])
```
