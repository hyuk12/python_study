## 전국 로컬 푸드 인증 정보 표준 데이터

- 품목명 top10 (품목별 카운트)
- dictionary 작성: {군단위 지역명 : [품목명]}
- dictionary 작성: {품목 : [군단위 지역명]}

```python
import numpy as np
import math


def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list=[]
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list))>1:
        for i in set(len_list):
            print(i, len_list.count(i))

def dist_np(p1, p2): #  [3,10]  [5,25]
    return math.sqrt(sum((p2-p1)**2))
```

```python
import requests, csv

url = "https://www.data.go.kr/download/15025455/standard.do?dataType=csv"
response = requests.get(url)

with open("data.csv", "wb") as f:
    f.write(response.content)


items = {}
item_count = {}
providers = {}
with open("data.csv", encoding="cp949") as f:
    reader = csv.DictReader(f)
    for row in reader:
        item_name = row["품목명"].split(" ")[0].split("(")[0].split("+")[0]
        provider_name = row["제공기관명"]
        
        if item_name not in item_count:
            item_count[item_name] = 0
        item_count[item_name] += 1
        
        if item_name not in items:
            items[item_name] = []
        items[item_name].append(provider_name)
        
        if provider_name not in providers:
            providers[provider_name] = []
        providers[provider_name].append(item_name)
        
item_count_list = [(item_name, count) for item_name, count in item_count.items()]
item_count_list.sort(key=lambda x: x[1], reverse=True)

print('품목명 top10\n')

for item_name, count in item_count_list[:10]:
    print(f"{item_name}: {count}")
print('품목 : 제공기관명\n')
    
for item_name, providers in items.items():
    print(f"{item_name}: {providers}")
print('제공기관명: 품목\n')  

for provider_name, items in providers.items():
    print(f"{provider_name}: {items}")
```

```python
import csv, requests
from collections import Counter

url = "https://www.data.go.kr/download/15025455/standard.do?dataType=csv"
response = requests.get(url)

with open("data.csv", "wb") as f:
    f.write(response.content)
    
item_names = []
provider_names = []
with open("data.csv", encoding="cp949") as f:
    reader = csv.DictReader(f)
    for row in reader:
        item_name = row["품목명"].split(" ")[0]
        item_names.append(item_name)
        provider_names.append(row["제공기관명"])

counter = Counter(item_names)
top_10 = counter.most_common(10)

for item_name, count in top_10:
    print(f"{item_name}: {count}")
```
