```python
import pandas as pd
import numpy as np


url = 'https://www.data.go.kr/download/15012891/standard.do?dataType=csv'
raw_df = pd.read_csv(url, encoding= 'cp949')
column_select = ['시설종류', '관할경찰서명', 'CCTV설치여부', 'CCTV설치대수', '제공기관명', '보호구역도로폭']
pd_data = raw_df[column_select]
pd_data['시도명'] = pd_data['제공기관명']
pd_data.drop('제공기관명', axis = 1, inplace = True)
```

```python
print(pd_data['시설종류'].unique())
print(pd_data['시설종류'].hasnans)

# 결측치 / 오류치 /이상치 제로
```

```python
print(pd_data['관할경찰서명'].unique())
print(pd_data['관할경찰서명'].hasnans)
# 결측치 / 오류치 /이상치 제로
```

```python
print(pd_data['CCTV설치여부'].unique())
print(pd_data['CCTV설치여부'].hasnans)
# 결측치 / 오류치 /이상치 제로
```

```python
print(pd_data['CCTV설치대수'].unique())
print(pd_data['CCTV설치대수'].hasnans)
np_filter1= (pd_data['CCTV설치여부'] == 'Y') & (pd_data['CCTV설치대수'].isna())
np_filter2= (pd_data['CCTV설치여부'] == 'N') & (pd_data['CCTV설치대수'].isna())

pd_data.loc[np_filter1, 'CCTV설치대수'] = 1
pd_data.loc[np_filter2, 'CCTV설치대수'] = 0
# 결측치 : nan 값중 설치가 되어있는 곳은 1개로 노설치는 0개로 바꿈
#/ 오류치 /이상치 제로

print(pd_data['CCTV설치대수'].unique())
print(pd_data['CCTV설치대수'].hasnans)
```

```python
print(pd_data['시도명'].unique())
print(pd_data['시도명'].hasnans)
# 결측치 / 오류치 /이상치 제로

def myfn2(x):
    if type(x) == type(''):
        return x.split()[0]
pd_data.loc[:, '시도명'] = pd_data.loc[:, '시도명'].apply(myfn2)
print(pd_data['시도명'].unique())
print(pd_data['시도명'].hasnans)
```

```python
print(pd_data['보호구역도로폭'].unique())
print(pd_data['보호구역도로폭'].hasnans)
# 결측치 / 오류치 /이상치 제로

def myfn3(x):
    if type(x) == type(''):
        if 'm' in x:
            t = x.split('m')[0]
            return str(t)
        elif '-' in x:
            t =  np.array(x.split('-')).astype(np.float64).mean()
            return str(t)
        elif '~' in x:
            t =  np.array(x.split('~')).astype(np.float64).mean()
            return str(t)
    
y = pd_data.loc[:, '보호구역도로폭'].apply(myfn3).astype(np.float64)
y = y.replace(np.nan, y.mean())
pd_data['보호구역도로폭'] = y

print(pd_data['보호구역도로폭'].unique())
print(pd_data['보호구역도로폭'].hasnans)
```

```python
pd_data.to_csv('전국어린이보호구역데이터.csv', encoding='cp949')
```
