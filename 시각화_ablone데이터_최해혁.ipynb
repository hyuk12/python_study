# 전복회를 주문하려는 사람!!

- Q1. 암수를 구분해서 주문하는 것이 이득인가?

- 범위는 데이터 전체 내에서 진행한다.
- 처음 받아오는 데이터는 숫자로 표현되어 있다 
    - 이 데이터를 보기좋게 일단은 영어로 컬럼명을 바꾼 후 내가 쓸 컬럼만 따로 빼서 한글로 변환후 진행한다.
    - 여기서 내가 쓸 데이터 시리즈는 성별, 전체 무게, 살점 무게, 내장 무게, 껍질 무게이다.
    
- 위의 데이터를 가지고 알 수 있는 것 !
    - 무조건 전체 무게가 크다고 양이 많지는 않을 것
    - 선호도에 따라 내장이 많은 것을 좋아할 수도 있고 살점이 많은 것을 좋아할 수도 있다.
    - 예를 들어 전복죽에 내장이 있는 것을 좋아하는 사람이 있고 없는 것을 좋아하는 사람이 있듯이!
    
    - 성별에 따라서 전체무게 평균을 내보고 그 평균 무게당 살점 무게 / 내장 무게 등을 내어서 상관관계를 따져본다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

font_fname = 'D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
pd_data = pd.read_csv(url, encoding='cp949')
pd_data.columns = ['sex', 'length', 'diameter', 'height', 'whole weight', 'shucked weight', 'viscera weight', 'shell weight', 'rings']
columns = ['sex','whole weight', 'shucked weight', 'viscera weight', 'shell weight' ]
```

```python
a1 = pd_data[columns]
```

```python
a1.columns = ['성별', '전체 무게', '살점 무게', '내장 무게', '껍질 무게']
```

```python
a1
```

### 성별에 대한 unique 값을 뽑았을 때 우리가 예상한 성별이외에 I라는 값이 있어서 제거하기로한다.

```python

a1.loc[a1['성별'] == 'I', '성별'] = np.nan
a1.dropna(axis=0, inplace=True)
```

### 성별에 따른 전체 무게, 살점무게, 내장 무게, 껍질 무게 등으로 그룹바이해준다

```python
b1 = a1.groupby('성별').mean()
b1
```

### 여기서 우리는 전체무게당 살점무게 내장무게 껍질무게 등을 알아볼 것이다.

```python
b1['살점무게/전체무게'] = b1['살점 무게']/ b1['전체 무게'] * 100
b1['내장무게/전체무게'] = b1['내장 무게']/ b1['전체 무게'] * 100
b1['껍질무게/전체무게'] = b1['껍질 무게']/ b1['전체 무게'] * 100
```

```python
c1 = b1.reset_index()
c1['살점무게/전체무게'][0]
```

## 시각화 계획

- 파이차트를 이용해서 간단하게 비율을 알아보려고한다.
- 그 비율퍼센트를 알아보고 암컷이 살점이 많은지 수컷이 많은지 알아본다.
- fig, axs를 설정하여 한화면에 두가지의 파이차트를 설계한다.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 20), subplot_kw={'aspect': 'equal'})
c1.pivot('성별', '전체 무게','살점 무게')

ratio1 = [c1['살점무게/전체무게'][0], c1['내장무게/전체무게'][0], c1['껍질무게/전체무게'][0]]
labels1 = ['살점무게/전체무게', '내장무게/전체무게', '껍질무게/전체무게']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
explode = [0.05, 0.05, 0.05]
axs[0].pie(ratio1, labels=labels1, autopct='%.1f%%',  startangle=260, explode = explode, shadow = True, wedgeprops=wedgeprops)
axs[0].legend(loc='center', bbox_to_anchor=(0.5, -0.3), fontsize=20)
axs[0].set_title('암컷의 전체무게당 비율', fontsize = 25)



ratio2 = [c1['살점무게/전체무게'][1], c1['내장무게/전체무게'][1], c1['껍질무게/전체무게'][1]]
labels2 = ['살점무게/전체무게', '내장무게/전체무게', '껍질무게/전체무게']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
explode = [0.05, 0.05, 0.05]
axs[1].pie(ratio2, labels=labels2, autopct='%.1f%%',  startangle=260, explode = explode, shadow = True, wedgeprops=wedgeprops)
axs[1].legend(loc='center', bbox_to_anchor=(0.5, -0.3), fontsize=20)
axs[1].set_title('수컷의 전체무게당 비율', fontsize = 25)

plt.tight_layout()
plt.show()
```

## 결론

- 처음 알고자 했던 것: 어떤 전복을 구매하는 것이 좋을까 (암컷/ 수컷)
    - 단면적으로 본다면 암컷의 전복이 평균적으로 살점무게가 높으니 좋다.
    
- 암컷과 수컷의 비교:
    - 둘다 같은 전체 무게라고 한다면
        비율로 따졌을 때 수컷을 사먹는 것이 더좋다고 생각이 든다.
    - 암컷과 수컷의 내장/살점/껍질의 비율이 전체무게를 기준으로 하였을 때 비슷하다.
        그 것을 토대로 퍼센트를 낸다면 암컷과 수컷의 비율로본다면 수컷의 살점 비율이 
        암컷의 살점 비율보다 살짝 높다.
    
- 전복죽에 내장을 넣는 것을 좋아하는 사람:
    이 경우에는 전체무게당 내장 비율이 높은 암컷을 구매하는 것이 맞아보인다.
