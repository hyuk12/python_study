```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_fname = 'D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

pd_data = pd.read_csv('이륜차신고현황.csv', encoding='cp949', header = [0, 1, 2])
pd_data.drop(('시도명(1)','시도명(1)', '시도명(1)'), axis=1, inplace=True)
pd_data.columns.names = ['시점', '소유', '크기']

a = pd_data.set_index(('시군구(1)', '시군구(1)', '시군구(1)'))
b = a.stack(level=0)
b.index.names = ['지역', '시점']
c = b.swaplevel(0,1)
d = c.stack(level = 0)
d = d.reset_index()
```

```python

d.insert(0, '년', d['시점'].apply(lambda x: str(x).split('.')[0]))

d.insert(1, '월', d['시점'].apply(lambda x: str(x).split('.')[1]))
# pd_data = pd_data.rename(columns={"시군구(1)": "지역"})


# pd_data = pd_data.drop(0, axis=0)

# # 첫 번째 행을 컬럼으로 빼서 컬럼 이름을 변경
# cols = pd_data.iloc[0]
# pd_data.columns = cols
# pd_data = pd_data.drop(1, axis=0)

# # 년, 월 컬럼을 따로 추출
# pd_data["년"], pd_data["월"] = pd_data["지역"].str.split(".", 1).str

# # 년, 월 컬럼을 앞으로 이동
# cols = pd_data.columns.tolist()
# cols = cols[-2:] + cols[:-2]
# pd_data = pd_data[cols]

# # 새로운 데이터 프레임에 저장
# pd_data.to_csv("new_data.csv", index=False)
```

```python
d.drop('소유', axis = 1, inplace = True)
```

```python
d
```

```python
e = d.groupby(['월', '지역']).sum()
```

```python
f = e.reset_index()
```

```python
# f['경형/소계'] = 100 * f['경형'] / f['소계'] 
# f['소형/소계'] = 100 * f['소형'] / f['소계'] 
# f['중형/소계'] = 100 * f['중형'] / f['소계'] 
# f['대형/소계'] = 100 * f['대형'] / f['소계'] 
fig = plt.figure(figsize=(20,10))
g = f.groupby(['소계', '지역']).mean()
h = g.reset_index()
h.pivot('지역', '소계')
sns.lineplot(data = h, x = '소계', y = '경형')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))

ax1 = plt.subplot(2,1,1)
ax1.plot(h['소계'], h['경형'], 'r--', label='경형 / 소계')
ax1.plot(h['소계'], h['소형'], 'b--', label='소형 / 소계')
ax1.plot(h['소계'], h['중형'], 'g--', label='중형 / 소계')
ax1.plot(h['소계'], h['대형'], 'y--', label='대형 / 소계')

# Add labels and title
ax1.set_xlabel('소계')
ax1.set_ylabel('size')
ax1.set_title('count size')

# Add legend
ax1.legend()

# Show plot

ax2 = plt.subplot(2,2,3)
sns.histplot(data = f, x ='지역', y='소계', hue='경형/소계', ax = ax2)
sns.histplot(data = f, x='지역', y='소계', hue='소형/소계',ax = ax2)
sns.histplot(data = f, x='지역', y='소계', hue='중형/소계',ax = ax2)
sns.histplot(data = f, x='지역', y='소계', hue='대형/소계',ax = ax2)

ax2.set_xlabel('소계')
ax2.set_ylabel('size')
ax2.set_title('count size')



# # Pie Chart for 2015
# market_sizes1 = [data_processing_and_management[0], data_construction_and_consulting[0], data_sales_and_provision[0]]

# industries = ['Data Processing and Management', 'Data Construction and Consulting', 'Data Sales and Provision']
# ax2 = plt.subplot(2,2,3)
# ax2.pie(market_sizes1, labels=industries, autopct='%1.1f%%')
# ax2.set_title('Market Size Composition in 2015')

# # Pie Chart for 2021
# market_sizes2 = [data_processing_and_management[6], data_construction_and_consulting[6], data_sales_and_provision[6]]

# industries = ['Data Processing and Management', 'Data Construction and Consulting', 'Data Sales and Provision']

# ax2 = plt.subplot(2,2,4)
# ax2.pie(market_sizes2, labels=industries, autopct='%1.1f%%')
# ax2.set_title('Market Size Composition in 2021')

# # Show plot

plt.show()
```
