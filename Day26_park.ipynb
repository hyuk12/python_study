```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_fname = 'D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

pd_data = pd.read_csv('park.csv.xls')
```

```python
pd_data

g1 = pd_data.groupby(['지자체명', '공원구분'])
g2 = g1.mean().reset_index().pivot('지자체명', '공원구분', '공원면적')
```

```python
g3 = g1.count().reset_index().pivot('지자체명', '공원구분', '공원면적')
g3.fillna(0.0)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(g2, annot = True, fmt='.0f')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(g3.fillna(0.0), annot = True, fmt='.0f')
plt.show()
```
